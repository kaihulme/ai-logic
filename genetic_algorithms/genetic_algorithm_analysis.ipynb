# Genetic Algorithm Analysis

Graphical analysis of genetic algorithm results.

## Setup

Dataframe and plotting tools:

```python
import pandas as pd
from matplotlib import pyplot as plt
```

Function to return .dat as a clean DataFrame:

```python
def get_clean_df(file_name): 

    file_name = "output/" + file_name
    column_names = ["gen_max", "min", "mean"]
    data = pd.read_csv(file_name, delimiter=",", names=column_names)

    data[["gen", "max"]] = data["gen_max"].str.split(':', 1, expand=True)
    del data["gen_max"]

    data["min"] = data["min"].str[5:].astype('float')
    data["max"] = data["max"].str[5:].astype('float')
    data["mean"] = data["mean"].str[6:].astype('float')
    data["gen"] = data["gen"].astype('int64')

    data = data[["gen","min","max","mean"]]
    return data
```

Get and clean data:

```python
file_name = "ga_output_max_gen_100_to_6400.dat"
data = get_clean_df(file_name)

print(data)
```

Function to plot min, max and mean with errorbars:

```python
def plot_dat(data, ee):

    fig, ax = plt.subplots(figsize=(12,8))

    ax.set_xlabel("Generation")
    ax.set_ylabel("Fitness")
    ax.legend(["min","max","mean"])
    ax.grid(axis='y', zorder=0)

    ax.plot(data['gen'], data['min'], c='orange')
    ax.plot(data['gen'], data['max'], c='blue')

    mean_std = data['mean'].std(axis=0)
    ax.errorbar(data['gen'], data['mean'], yerr=mean_std, errorevery=ee, c='black', ecolor='grey')

    plt.show()
```

Plot data:

```python
plot_dat(data)
```
